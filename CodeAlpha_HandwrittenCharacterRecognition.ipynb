```python
"""
CodeAlpha Internship — Task 3: Handwritten Character Recognition
Author: [Your Name]
Description: CNN-based recognition of handwritten digits (MNIST) and characters (EMNIST).
             Includes training, evaluation, visualization, and a simple inference demo.
"""

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, regularizers
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.metrics import classification_report, confusion_matrix
import warnings
warnings.filterwarnings('ignore')

# ─────────────────────────────────────────────
# CONFIG
# ─────────────────────────────────────────────
SEED = 42
tf.random.set_seed(SEED)
np.random.seed(SEED)

IMG_SIZE  = 28
N_CLASSES_MNIST  = 10
N_CLASSES_EMNIST = 26  # letters A–Z
BATCH_SIZE = 128
EPOCHS     = 30


# ─────────────────────────────────────────────
# 1. BUILD CNN MODEL
# ─────────────────────────────────────────────
def build_cnn(input_shape=(28, 28, 1), n_classes=10, name='CNN'):
    """
    Deep CNN with:
      - 3 convolutional blocks (Conv → BN → ReLU → MaxPool → Dropout)
      - Fully connected head with L2 regularisation
      - Softmax output
    """
    inputs = keras.Input(shape=input_shape, name='image')

    # Block 1
    x = layers.Conv2D(32, 3, padding='same', kernel_regularizer=regularizers.l2(1e-4))(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    x = layers.Conv2D(32, 3, padding='same', kernel_regularizer=regularizers.l2(1e-4))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    x = layers.MaxPooling2D(2)(x)
    x = layers.Dropout(0.25)(x)

    # Block 2
    x = layers.Conv2D(64, 3, padding='same', kernel_regularizer=regularizers.l2(1e-4))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    x = layers.Conv2D(64, 3, padding='same', kernel_regularizer=regularizers.l2(1e-4))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    x = layers.MaxPooling2D(2)(x)
    x = layers.Dropout(0.25)(x)

    # Block 3
    x = layers.Conv2D(128, 3, padding='same', kernel_regularizer=regularizers.l2(1e-4))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    x = layers.MaxPooling2D(2)(x)
    x = layers.Dropout(0.4)(x)

    # Dense head
    x = layers.Flatten()(x)
    x = layers.Dense(256, kernel_regularizer=regularizers.l2(1e-4))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    x = layers.Dropout(0.5)(x)
    outputs = layers.Dense(n_classes, activation='softmax', name='predictions')(x)

    model = keras.Model(inputs, outputs, name=name)
    return model


# ─────────────────────────────────────────────
# 2. DATA AUGMENTATION
# ─────────────────────────────────────────────
def get_augmenter():
    return ImageDataGenerator(
        rotation_range=10,
        zoom_range=0.1,
        width_shift_range=0.1,
        height_shift_range=0.1,
        shear_range=0.1,
        fill_mode='nearest'
    )


# ─────────────────────────────────────────────
# 3. LOAD & PREPROCESS MNIST
# ─────────────────────────────────────────────
def load_mnist():
    print("\n[MNIST] Loading digits dataset...")
    (X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()

    X_train = X_train[..., np.newaxis] / 255.0
    X_test  = X_test[..., np.newaxis]  / 255.0

    y_train_cat = keras.utils.to_categorical(y_train, N_CLASSES_MNIST)
    y_test_cat  = keras.utils.to_categorical(y_test,  N_CLASSES_MNIST)

    labels = [str(i) for i in range(10)]
    print(f"  Train: {X_train.shape}, Test: {X_test.shape}")
    return X_train, y_train, y_train_cat, X_test, y_test, y_test_cat, labels


# ─────────────────────────────────────────────
# 4. LOAD & PREPROCESS EMNIST (Letters)
# ─────────────────────────────────────────────
def load_emnist():
    """
    EMNIST Letters via TensorFlow Datasets.
    Falls back to a synthetic subset if tfds is unavailable.
    """
    print("\n[EMNIST] Loading letters dataset...")
    try:
        import tensorflow_datasets as tfds
        ds_train, ds_test = tfds.load(
            'emnist/letters',
            split=['train', 'test'],
            as_supervised=True
        )
        def preprocess(img, label):
            img = tf.cast(img, tf.float32) / 255.0
            img = tf.image.transpose(img)          # EMNIST needs transpose
            label = label - 1                       # Labels are 1–26 → 0–25
            return img, label

        AUTOTUNE = tf.data.AUTOTUNE
        train_ds = ds_train.map(preprocess, AUTOTUNE).cache().shuffle(100000).batch(BATCH_SIZE).prefetch(AUTOTUNE)
        test_ds  = ds_test.map(preprocess, AUTOTUNE).cache().batch(BATCH_SIZE).prefetch(AUTOTUNE)

        # Extract numpy arrays for sklearn metrics
        X_test_np = np.concatenate([x.numpy() for x, _ in test_ds], axis=0)
        y_test_np = np.concatenate([y.numpy() for _, y in test_ds], axis=0)
        labels = [chr(ord('A') + i) for i in range(26)]
        return train_ds, test_ds, X_test_np, y_test_np, labels, 'tfds'

    except Exception as e:
        print(f"  tfds unavailable ({e}), using MNIST as proxy for demo.")
        # Use MNIST digits as fallback demonstration
        (X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
        X_train = X_train[..., np.newaxis] / 255.0
        X_test  = X_test[..., np.newaxis]  / 255.0
        # Limit to first 10 classes (0–9) as proxy
        labels = [str(i) for i in range(10)]
        return X_train, y_train, X_test, y_test, labels, 'numpy'


# ─────────────────────────────────────────────
# 5. TRAINING FUNCTION
# ─────────────────────────────────────────────
def train_model(model, X_train, y_train_cat, X_test, y_test_cat,
                task_name='MNIST', use_augmentation=True):
    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=1e-3),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    model.summary()

    callbacks = [
        EarlyStopping(patience=6, restore_best_weights=True, verbose=1),
        ReduceLROnPlateau(patience=3, factor=0.5, min_lr=1e-6, verbose=1),
        ModelCheckpoint(f'best_{task_name.lower()}_model.keras',
                        save_best_only=True, verbose=0)
    ]

    if use_augmentation:
        aug = get_augmenter()
        aug.fit(X_train)
        history = model.fit(
            aug.flow(X_train, y_train_cat, batch_size=BATCH_SIZE),
            epochs=EPOCHS,
            validation_data=(X_test, y_test_cat),
            callbacks=callbacks,
            verbose=1
        )
    else:
        history = model.fit(
            X_train, y_train_cat,
            batch_size=BATCH_SIZE,
            epochs=EPOCHS,
            validation_split=0.1,
            callbacks=callbacks,
            verbose=1
        )
    return history


# ─────────────────────────────────────────────
# 6. EVALUATION & VISUALIZATION
# ─────────────────────────────────────────────
def evaluate_model(model, X_test, y_test, labels, task_name):
    y_prob = model.predict(X_test, verbose=0)
    y_pred = np.argmax(y_prob, axis=1)

    print(f"\n{'='*55}")
    print(f"EVALUATION — {task_name}")
    print(f"{'='*55}")
    print(classification_report(y_test, y_pred, target_names=labels))

    cm = confusion_matrix(y_test, y_pred)
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))

    # Confusion Matrix
    sns.heatmap(cm, annot=(len(labels) <= 15), fmt='d', cmap='Blues',
                xticklabels=labels, yticklabels=labels, ax=axes[0],
                annot_kws={'size': 7})
    axes[0].set_title(f'{task_name} — Confusion Matrix')
    axes[0].set_xlabel('Predicted')
    axes[0].set_ylabel('Actual')
    if len(labels) > 15:
        axes[0].set_xticklabels(labels, fontsize=7)
        axes[0].set_yticklabels(labels, fontsize=7)

    # Per-class accuracy
    per_class_acc = cm.diagonal() / cm.sum(axis=1)
    axes[1].bar(labels, per_class_acc, color='steelblue', edgecolor='white')
    axes[1].set_title(f'{task_name} — Per-Class Accuracy')
    axes[1].set_ylim(0.8, 1.0)
    axes[1].set_xlabel('Class')
    axes[1].set_ylabel('Accuracy')

    plt.tight_layout()
    plt.savefig(f'eval_{task_name.lower().replace(" ","_")}.png', dpi=150)
    plt.show()
    plt.close()


def plot_training(history, task_name):
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    axes[0].plot(history.history['accuracy'],     label='Train')
    axes[0].plot(history.history['val_accuracy'], label='Val')
    axes[0].set_title(f'{task_name} — Accuracy')
    axes[0].legend()

    axes[1].plot(history.history['loss'],     label='Train')
    axes[1].plot(history.history['val_loss'], label='Val')
    axes[1].set_title(f'{task_name} — Loss')
    axes[1].legend()

    plt.tight_layout()
    plt.savefig(f'training_{task_name.lower().replace(" ","_")}.png', dpi=150)
    plt.show()
    plt.close()


def visualize_predictions(model, X_test, y_test, labels, task_name, n=20):
    """Show a grid of predictions with correct (green) / wrong (red) borders."""
    indices = np.random.choice(len(X_test), n, replace=False)
    y_prob  = model.predict(X_test[indices], verbose=0)
    y_pred  = np.argmax(y_prob, axis=1)

    cols = 10
    rows = n // cols
    fig, axes = plt.subplots(rows, cols, figsize=(20, rows * 2.2))
    for i, (ax, idx) in enumerate(zip(axes.flatten(), indices)):
        ax.imshow(X_test[idx].squeeze(), cmap='gray')
        pred  = labels[y_pred[i]]
        truth = labels[y_test[idx]]
        color = 'green' if y_pred[i] == y_test[idx] else 'red'
        ax.set_title(f'P:{pred} T:{truth}', fontsize=8, color=color)
        ax.axis('off')
        for spine in ax.spines.values():
            spine.set_edgecolor(color)
            spine.set_linewidth(2)
    plt.suptitle(f'{task_name} — Sample Predictions', y=1.02, fontsize=13)
    plt.tight_layout()
    plt.savefig(f'predictions_{task_name.lower().replace(" ","_")}.png', dpi=150, bbox_inches='tight')
    plt.show()
    plt.close()
    print(f"[Viz] Predictions plot saved for {task_name}.")


# ─────────────────────────────────────────────
# 7. MAIN
# ─────────────────────────────────────────────
def main():
    print("=" * 55)
    print("  CodeAlpha — TASK 3: Handwritten Character Recognition")
    print("=" * 55)

    # ── MNIST Digits ──────────────────────────
    X_train, y_train, y_train_cat, X_test, y_test, y_test_cat, labels_mnist = load_mnist()

    model_mnist = build_cnn(input_shape=(IMG_SIZE, IMG_SIZE, 1),
                            n_classes=N_CLASSES_MNIST,
                            name='CNN_MNIST')
    history_mnist = train_model(model_mnist, X_train, y_train_cat,
                                X_test, y_test_cat, task_name='MNIST')

    plot_training(history_mnist, 'MNIST')
    evaluate_model(model_mnist, X_test, y_test, labels_mnist, 'MNIST')
    visualize_predictions(model_mnist, X_test, y_test, labels_mnist, 'MNIST')

    # ── EMNIST Letters ────────────────────────
    emnist_data = load_emnist()
    *train_data, labels_em, mode = emnist_data

    if mode == 'tfds':
        train_ds, test_ds, X_test_em, y_test_em, labels_em, _ = emnist_data
        model_em = build_cnn(input_shape=(IMG_SIZE, IMG_SIZE, 1),
                             n_classes=N_CLASSES_EMNIST,
                             name='CNN_EMNIST')
        model_em.compile(
            optimizer=keras.optimizers.Adam(1e-3),
            loss='sparse_categorical_crossentropy',
            metrics=['accuracy']
        )
        callbacks = [
            EarlyStopping(patience=6, restore_best_weights=True, verbose=1),
            ReduceLROnPlateau(patience=3, factor=0.5, verbose=1)
        ]
        history_em = model_em.fit(
            train_ds, epochs=EPOCHS,
            validation_data=test_ds,
            callbacks=callbacks, verbose=1
        )
        y_prob = model_em.predict(test_ds, verbose=0)
        y_pred = np.argmax(y_prob, axis=1)

    else:
        # Fallback: retrain on MNIST as proxy
        # train_data = [X_train, y_train, X_test, y_test]  (4 items, labels_em already unpacked above)
        X_tr_em, y_tr_em, X_te_em, y_te_em = train_data
        y_tr_cat = keras.utils.to_categorical(y_tr_em, 10)
        y_te_cat = keras.utils.to_categorical(y_te_em, 10)
        model_em = build_cnn(input_shape=(IMG_SIZE, IMG_SIZE, 1),
                             n_classes=10, name='CNN_EMNIST_proxy')
        history_em = train_model(model_em, X_tr_em, y_tr_cat,
                                 X_te_em, y_te_cat, task_name='EMNIST_proxy')
        X_test_em, y_test_em = X_te_em, y_te_em

    plot_training(history_em, 'EMNIST')
    evaluate_model(model_em, X_test_em, y_test_em, labels_em, 'EMNIST')
    visualize_predictions(model_em, X_test_em, y_test_em, labels_em, 'EMNIST')

    print("\n✅ All outputs saved. Upload to GitHub as: CodeAlpha_HandwrittenCharacterRecognition")


if __name__ == '__main__':
    main()
```
